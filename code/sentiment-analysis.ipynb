```python
import pandas as pd
import os
```

```python
# Define a function to calculate sentiment percentages and totals
def calculate_sentiment_percentages(folder_path):
    # Initialize a list to store the results
    results = []

    # Iterate over each CSV file in the folder
    for file_name in os.listdir(folder_path):
        if file_name.endswith('.csv'):
            # Read the CSV file into a DataFrame
            df = pd.read_csv(os.path.join(folder_path, file_name))

            # Calculate the total number of sentiments
            total_sentiments = len(df)

            # Calculate the number of each type of sentiment
            positive_count = len(df[df['sentiment'] == 'positive'])
            negative_count = len(df[df['sentiment'] == 'negative'])
            neutral_count = len(df[df['sentiment'] == 'neutral'])

            # Calculate the percentage of each type of sentiment
            positive_percentage = (positive_count / total_sentiments) * 100
            negative_percentage = (negative_count / total_sentiments) * 100
            neutral_percentage = (neutral_count / total_sentiments) * 100

            # Append the results to the list
            results.append({
                'file_name': file_name,
                'positive_percentage': positive_percentage,
                'negative_percentage': negative_percentage,
                'neutral_percentage': neutral_percentage,
                'positive_total': positive_count,
                'negative_total': negative_count,
                'neutral_total': neutral_count,
                'total_num_of_sentiment': total_sentiments
            })

    # Convert the results list to a DataFrame
    results_df = pd.DataFrame(results)

    # Return the path to the new CSV file
    return results_df

# # Assuming the CSV files are in a folder called 'sentiment_data' in the current directory
# folder_path = 'sentiment_data'

# # Call the function and print the path to the new CSV file
# new_csv_path = calculate_sentiment_percentages(folder_path)
# print(f"The sentiment summary has been saved to {new_csv_path}")
```
